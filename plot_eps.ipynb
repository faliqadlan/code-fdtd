```python
import pandas as pd

# Load the CSV file using pandas and display the table
import pandas as pd

dataNK = pd.read_csv("n-k-data.csv")

# Display the table
dataNK
```

```python
# set data for wavelengt and n k materials
wavelengths = dataNK["Wavelength (nm)"]
wavelengthsUm = [i / 1000 for i in wavelengths]
freqs = [1000 / (i) for i in wavelengths]
auN = dataNK["Au_n"]
auK = dataNK["Au_k"]
spiroN = dataNK["Spiro_n"]
spiroK = dataNK["Spiro_k"]
perovN = dataNK["Perov_n"]
perovK = dataNK["Perov_k"]
tio2N = dataNK["TiO2_n"]
tio2K = dataNK["TiO2_k"]
```

```python
epsPerov = [complex(perovN[i], perovK[i])**2 for i in range(len(perovN))]

# Plot the data epsilon of perovskite
import matplotlib.pyplot as plt

plt.plot(wavelengthsUm, [i.real for i in epsPerov], label="Real")
plt.plot(wavelengthsUm, [i.imag for i in epsPerov], label="Imaginary")
plt.plot(wavelengthsUm, perovN, label="n")
plt.plot(wavelengthsUm, perovK, label="k")
plt.xlabel("Wavelength (um)")
plt.ylabel("Epsilon")
plt.legend()
plt.title("Perovskite Epsilon")
plt.show()
```

```python
def plotEps(wavelength, nArray, kArray):
    eps = [complex(nArray[i], kArray[i]) ** 2 for i in range(len(kArray))]
    print(eps)
    plt.scatter(wavelength, [i.real for i in eps], label="Real", s=0.5)
    plt.scatter(wavelength, [i.imag for i in eps], label="Imaginary", s=0.5)
    plt.xlabel("Wavelength (nm)")
    plt.ylabel("Epsilon")
    plt.legend()
    plt.show()

plotEps(wavelengths, perovN, perovK)
```
